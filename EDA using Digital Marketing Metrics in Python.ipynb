# Exploratory Data Analysis(EDA) using Digital Marketing Metrics in Python 

### Overview:

Without a competent and timely analysis of the KPIs and 'Digital Marketing Metrics'  the digital marketing, Campaigns, web analytics lose their meaning.
There are various types of Digital Marketing Metrics, which is very helpful for any Business, here I've covered 2 of them and perform Exploratory Data Analysis. So, these are -

- Return on Invest(ROI)
- Customer Lifetime Value(CLV)

'Return on Investment' is a performance metric that’s used to evaluate the efficiency of a particular investment.

'Customer lifetime Value' analysis is used to estimate the total value of customers to the business over the lifetime of their relationship. It helps businesses make data-driven decisions on how to allocate their resources and improve their customer relationships in 'Add Campaigns'.

##### Some Important points regarding ROI & CLV:

- ROI is the queen of KPIs, even among those who have never heard about analytics! Return on investment is a performance metric that’s used to evaluate the efficiency of a particular investment.


- Customer lifetime value analysis is used to estimate the total value of customers to the business over the lifetime of their relationship. It helps companies determine how much to invest in customer acquisition and retention, as well as identify the most valuable customers to prioritize for retention efforts. By analyzing customer lifetime value, companies can identify the most effective marketing channels and campaigns for acquiring high-value customers, as well as develop targeted retention strategies to keep those customers engaged and loyal.

#### 1. Importing Necessary Libraries:

```python
# importing necessary libraries :

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 2. Loading the Data into Pandas Dataframe: 

```python
data = pd.read_csv(r"C:\Users\HP\Desktop\xmassbesra\Python_Project\Project\Project2\customer_acquisition_data.csv")
```

#### 3. Printing the First 5 rows of data:

```python
data.head()
```

#### 4. Visualizing the Distribution of Acquisition Cost and Revenue generated by the Customer using Histograms:

```python
# Distribution of Acquisition Cost:

sns.histplot(data=data, x="cost", bins=20)
sns.set_style('whitegrid')
sns.set(rc={'figure.figsize':(10,6)})
plt.title('Distribution of Acquisition Cost')
plt.show()
```

```python
# Distribution of Revenue:

sns.histplot(data=data, x="revenue", bins=20)
sns.set_style('whitegrid')
sns.set(rc={'figure.figsize':(10,6)})
plt.title('Distribution of Revenue')
plt.show()
```

#### 5. Comparing the cost of acquisition across different channels and identify the most and least profitable channels:

```python
# Customer Acquisition Cost by Channel:

cost_by_channel = data.groupby('channel')['cost'].mean().reset_index()

sns.barplot(data=cost_by_channel, x='channel', y='cost')
sns.set_style('whitegrid')
sns.set(rc={'figure.figsize':(10,6)})
plt.title('Customer Acquisition Cost by Channel')
plt.show()
```

```python
# Conversion Rate by Channel:

conversion_by_channel = data.groupby('channel')['conversion_rate'].mean().reset_index()

sns.barplot(data=conversion_by_channel, x='channel', y='conversion_rate')
sns.set_style('whitegrid')
sns.set(rc={'figure.figsize':(10,6)})
plt.title('Conversion Rate by Channel')
plt.show()
```

#### 6.  The most and least profitable channels in terms of generating revenue:

```python
revenue_by_channel = data.groupby('channel')['revenue'].sum().reset_index()

plt.pie(revenue_by_channel['revenue'], labels=revenue_by_channel['channel'], 
        autopct='%1.1f%%', pctdistance=0.8, colors=sns.color_palette('pastel'), 
        wedgeprops={'width': 0.6, 'edgecolor': 'w'})

plt.title('Total Revenue by Channel')
plt.show()
```

#### 7. Judging the Ad-Campaigns by Digital Marketing Metrices:

###### a) Return on Invest(ROI) :

```python
# Return on Invest(ROI) by channel :

data['roi'] = data['revenue'] / data['cost']
roi_by_channel = data.groupby('channel')['roi'].mean().reset_index()

sns.barplot(data=roi_by_channel, x='channel', y='roi')
sns.set_style('whitegrid')
sns.set(rc={'figure.figsize':(10,6)})
plt.title('Return on Investment (ROI) by Channel')
plt.show()
```

##### b) Customer Lifetime Value (CLV):

- CLTV = (revenue – cost) * conversion_rate / cost

```python
# Customer Lifetime Value:

data['cltv'] = (data['revenue'] - data['cost']) * data['conversion_rate'] / data['cost']
channel_cltv = data.groupby('channel')['cltv'].mean().reset_index()

sns.barplot(data=channel_cltv, x='channel', y='cltv', hue='channel')
sns.set_style('whitegrid')
sns.set(rc={'figure.figsize':(10,6)})
plt.title('Customer Lifetime Value by Channel')
plt.xlabel('Channel')
plt.ylabel('CLTV')
plt.show()
```

#### 8. Comparing the CLV distributions of the social media and referral channels:

```python
subset = data.loc[data['channel'].isin(['social media', 'referral'])]

sns.boxplot(data=subset, x='channel', y='cltv', palette='Set2')
sns.set_style('whitegrid')
sns.set(rc={'figure.figsize':(5,5)})
plt.title('CLTV Distribution by Channel')
plt.xlabel('Channel')
plt.ylabel('CLTV')
plt.show()
```
